# A Whale of a Portfolio

```python
import pandas as pd
import numpy as np
import datetime as dt
from pathlib import Path
import hvplot
```

## Data Cleaning
### Read the Whale Portfolio daily returns and clean the data

```python
whale_returns_csv = Path("Resources/whale_returns.csv")
```

```python
whale = pd.read_csv(
    whale_returns_csv, index_col="Date", infer_datetime_format=True, parse_dates=True)
whale.head()
```

```python
 # Checking for null
whale.isnull().sum()
```

```python
 # Drop nulls
whale.dropna(inplace = True)
whale.isnull().sum()
```

## Algorithimic Daily Returns
### Read the algorithmic daily returns and clean the data

```python
 # Reading algorithmic returns
algo_returns_csv = Path("Resources/algo_returns.csv")
algo = pd.read_csv(
    algo_returns_csv, index_col="Date", infer_datetime_format=True, parse_dates=True)
algo.head()
```

```python
 # Checking for null
algo.isnull().sum()
```

```python
 # Drop nulls
algo.dropna(inplace = True)
algo.isnull().sum()
```

## S&P 500 Returns
### Read the S&P500 Historic Closing Prices and create a new daily returns DataFrame from the data.

```python
 # Reading algorithmic returns
sp500_returns_csv = Path("Resources/sp500_history.csv")
sp500 = pd.read_csv(
    sp500_returns_csv, index_col="Date", infer_datetime_format=True, parse_dates=True)
sp500.head()
```

```python
 # Check Data Types
sp500.dtypes
```

```python
# Fix Data Types
sp500[['Close']] = sp500[['Close']].replace('[\$,]','',regex=True).astype(float)
sp500.dtypes
```

```python
 # Calculate Daily Returns
sp500_returns = sp500.pct_change()
sp500_returns.head()
```

```python
 # Drop nulls
sp500_returns.dropna(inplace = True)
sp500_returns.head()
```

```python
 # Rename Column
sp500_renamed = sp500_returns.rename(columns = {
'Close' : "S&P 500"})
sp500_renamed.head()
```

## Combine Whale, Algorithmic, and S&P 500 Returns

```python
 # Concatenate all DataFrames into a single DataFrame
combined_portfolio = pd.concat([whale, algo, sp500_renamed], axis="columns", join="inner")
combined_portfolio.head()
```

## Portfolio Analysis
 In this section, you will calculate and visualize performance and risk metrics for the portfolios. 
### Performance
<br> Calculate and Plot the daily returns and cumulative returns. Does any portfolio outperform the S&P 500?

```python
#Plot daily returns
daily_returns = combined_portfolio
daily_returns.plot(figsize=(20, 10), title='Daily Returns')
```

```python
 # Plot cumulative returns
cumulative_returns = (1 + daily_returns).cumprod()
cumulative_returns.plot(figsize=(20, 10), title='Cumulative Returns')    
```

### Risk
Determine the risk of each portfolio: <br>
1. Create a box plot for each portfolio. 
2. Calculate the standard deviation for all portfolios
3. Determine which portfolios are riskier than the S&P 500
4. Calculate the Annualized Standard Deviation

```python
# Box plot to visually show risk
Portfolio_Risk = daily_returns.boxplot(figsize=(20, 10), grid=False, fontsize=10)
```

```python
# Daily Standard Deviations
# Calculate the standard deviation for each portfolio. 
risk = daily_returns.std()
risk["S&P 500"]

# Which portfolios are riskier than the S&P 500?
```

```python
 # Determine which portfolios are riskier than the S&P 500
risk > risk["S&P 500"]
```

```python
# Calculate the annualized standard deviation (252 trading days)
annual_risk = risk * np.sqrt(252)
annual_risk
```
