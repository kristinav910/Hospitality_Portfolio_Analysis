# A Whale of a Portfolio

```python
import pandas as pd
import numpy as np
import datetime as dt
from pathlib import Path
import hvplot
```

## Data Cleaning
### Read the Whale Portfolio daily returns and clean the data

```python
whale_returns_csv = Path("Resources/whale_returns.csv")
```

```python
whale = pd.read_csv(
    whale_returns_csv, index_col="Date", infer_datetime_format=True, parse_dates=True)
whale.head()
```

```python
 # Checking for null
whale.isnull().sum()
```

```python
 # Drop nulls
whale.dropna(inplace = True)
whale.isnull().sum()
```

## Algorithimic Daily Returns
### Read the algorithmic daily returns and clean the data

```python
 # Reading algorithmic returns
algo_returns_csv = Path("Resources/algo_returns.csv")
algo = pd.read_csv(
    algo_returns_csv, index_col="Date", infer_datetime_format=True, parse_dates=True)
algo.head()
```

```python
 # Checking for null
algo.isnull().sum()
```

```python
 # Drop nulls
algo.dropna(inplace = True)
algo.isnull().sum()
```

## S&P 500 Returns
### Read the S&P500 Historic Closing Prices and create a new daily returns DataFrame from the data.

```python
 # Reading algorithmic returns
sp500_returns_csv = Path("Resources/sp500_history.csv")
sp500 = pd.read_csv(
    sp500_returns_csv, index_col="Date", infer_datetime_format=True, parse_dates=True)
sp500.head()
```

```python
 # Check Data Types
sp500.dtypes
```

```python
# Fix Data Types
sp500[['Close']] = sp500[['Close']].replace('[\$,]','',regex=True).astype(float)
sp500.dtypes
```

```python
 # Calculate Daily Returns
sp500_returns = sp500.pct_change()
sp500_returns.head()
```

```python
 # Drop nulls
sp500_returns.dropna(inplace = True)
sp500_returns.head()
```

```python
 # Rename Column
sp500_renamed = sp500_returns.rename(columns = {
'Close' : "S&P 500"})
sp500_renamed.head()
```

## Combine Whale, Algorithmic, and S&P 500 Returns

```python
 # Concatenate all DataFrames into a single DataFrame
combined_portfolio = pd.concat([whale, algo, sp500_renamed], axis="columns", join="inner")
combined_portfolio.head()
```

 #### Calculate Beta for a single portfolio (Berkshire Hathaway) compared to the total market (S&P 500)

```python
#Calculate covariance

BH_covariance = daily_returns['BERKSHIRE HATHAWAY INC'].cov(daily_returns['S&P 500'])
BH_covariance
```

```python
#calculate variance for S&P500 
variance = daily_returns['S&P 500'].var()
variance
```

```python
#calculate the beta for BH
BH_beta = BH_covariance / variance
BH_beta
```

```python
# Calculate rolling covariance of BH vs. S&P 500 and plot the data
# Calculate rolling covariance of BH 
rolling_BH_covariance = daily_returns['BERKSHIRE HATHAWAY INC'].rolling(window = 30).cov(daily_returns['S&P 500'])

# Calculate 30-day rolling variance of S&P 500
rolling_variance = daily_returns['S&P 500'].rolling(window = 30).var()

rolling_beta = rolling_BH_covariance / rolling_variance
rolling_beta.plot(figsize=(20, 10), title='Berkshire Hathaway Inc. Beta')
```

```python
# Calculate a rolling window using the exponentially weighted moving average. 
whale_expanded = daily_returns.ewm(halflife = 21).std().plot(figsize=(20, 10), title='Exponentially Weighted Average')
```

### Sharpe Ratios
In reality, investment managers and thier institutional investors look at the ratio of return-to-risk, and not just returns alone. (After all, if you could invest in one of two portfolios, each offered the same 10% return, yet one offered lower risk, you'd take that one, right?)
1. Using the daily returns, calculate and visualize the Sharpe ratios using a bar plot.
2. Determine whether the algorithmic strategies outperform both the market (S&P 500) and the whales portfolios.

```python
 # Calculate annualized Sharpe Ratios
sharpe_ratios = (daily_returns.mean() * 252) / (daily_returns.std() * np.sqrt(252))
sharpe_ratios
```

```python
sharpe_ratios.plot.bar(rot = 90, title = "Sharpe Ratios")
```
